```python
import numpy as np
```

```python
def sherman_morrison(v, w, A1, A2, A3):
    n = v.size
    matrix = np.identity(n) + A1 * np.outer(w, w) + A2 * np.outer(w, v) + A3 * np.outer(v, v)
    inv_numeric = np.linalg.inv(matrix)
    inv_analytic = np.identity(n) - (A1 * np.outer(w, w) + A2 * np.outer(w, v) + A3 * np.outer(v, v))/(1 + A1 * np.sum(w**2) + A2 * np.sum(v*w) + A3 * np.sum(v**2))
    return inv_numeric, inv_analytic
```

```python
v = np.arange(1,10)
w = np.ones(v.shape)
```

```python
num, anal = sherman_morrison(v, w, 1, 1, 0)
```
