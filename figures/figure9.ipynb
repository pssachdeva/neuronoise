```python
import matplotlib.pyplot as plt

from nd.plot_utils import (colors,
                           plot_scatter_linear,
                           plot_scatter_quadratic)

%matplotlib inline
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

plot_scatter_linear(N=1000, s=10, kv=1, kw=1, n_trials=1000, fax=(fig, axes[0]), color=colors[0])
plot_scatter_linear(N=1000, s=20, kv=1, kw=4, n_trials=1000, fax=(fig, axes[0]), color=colors[1])
plot_scatter_linear(N=1000, s=30, kv=1, kw=8, n_trials=1000, fax=(fig, axes[0]), color=colors[2])
lgd = axes[0].legend(prop={'size' : 30}, loc=4)
lgd.get_frame().set_edgecolor('k')
axes[0].tick_params(labelsize=20)
axes[0].set_xlabel(r'Neuron 1 Response', fontsize=35)
axes[0].set_ylabel(r'Neuron 2 Response', fontsize=35)
axes[0].set_xlim([0, 60])
axes[0].set_ylim([0, 60])
axes[0].set_xticks([])
axes[0].set_xticklabels([])
axes[0].set_yticks([])
axes[0].set_yticklabels([])

plot_scatter_quadratic(N=1000, s=10, kv=1, kw=1, n_trials=1000, fax=(fig, axes[1]), color=colors[0])
plot_scatter_quadratic(N=1000, s=20, kv=1, kw=4, n_trials=1000, fax=(fig, axes[1]), color=colors[1])
plot_scatter_quadratic(N=1000, s=30, kv=1, kw=8, n_trials=1000, fax=(fig, axes[1]), color=colors[2])
lgd = axes[1].legend(prop={'size' : 30})
lgd.get_frame().set_edgecolor('k')
axes[1].tick_params(labelsize=20)
axes[1].set_xlabel(r'Neuron 1 Response', fontsize=35)
axes[1].set_ylabel(r'Neuron 2 Response', fontsize=35)
axes[1].set_xlim([0, 3000])
axes[1].set_ylim([0, 3000])
axes[1].set_xticks([])
axes[1].set_xticklabels([])
axes[1].set_yticks([])
axes[1].set_yticklabels([])

# Subplot labels
axes[0].text(
    x=-0.1, y=1.1, s=r'\textbf{a}',
    fontsize=40,
    horizontalalignment='center',
    verticalalignment='center',
    transform=axes[0].transAxes)

axes[1].text(
    x=-0.1, y=1.1, s=r'\textbf{b}',
    fontsize=40,
    horizontalalignment='center',
    verticalalignment='center',
    transform=axes[1].transAxes)

for ax in axes.ravel():
    ax.tick_params(labelsize=30)

plt.subplots_adjust(wspace=0.3)
plt.savefig('figure9.pdf', bbox_inches='tight')
```
